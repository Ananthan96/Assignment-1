# Reading the excel file

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv(r'C:\Users\Ananthan\Downloads\StudentsPerformance.csv')
```

# Count of male and female 

```python
data['gender'].value_counts()
```

# Parental Level Of Education

```python
plot1=data['parental level of education'].value_counts().plot(kind='bar',figsize=(6,6),title='Parental Level Of Education')
plt.xlabel('Education Level')
plt.ylabel('Number Of Parents')
```

From the plot it is seen that majority of them have some college education or an associates degree. 
A few even have bachelors degree and masters degree.
The remaining parents all have atleast high school level of education

# Avg score by gender for each test

```python
plt.figure()
plt.subplot(1,3,1)
plot2=data.groupby('gender')['math score'].mean().plot(kind='bar',figsize=(12,4),title='Avg math score by gender')
plt.xlabel('Gender')
plt.ylabel('Avg Score')
plt.subplot(1,3,2)
plot2=data.groupby('gender')['reading score'].mean().plot(kind='bar',figsize=(12,4),title='Avg reading score by gender')
plt.xlabel('Gender')
plt.ylabel('Avg Score')
plt.subplot(1,3,3)
plot2=data.groupby('gender')['writing score'].mean().plot(kind='bar',figsize=(12,4),title='Avg writing score by gender')
plt.xlabel('Gender')
plt.ylabel('Avg Score')
```

From the plot it is seen that males performed better in maths whereas females performed better in reading and writing

# Avg score by test preparation course

```python
plt.figure()
plt.subplot(1,3,1)
plot3=data.groupby('test preparation course')['math score'].mean().plot(kind='bar',figsize=(12,4),title='Avg math score by TPC')
plt.xlabel('Test Preparation Course')
plt.ylabel('Avg Score')
plt.subplot(1,3,2)
plot3=data.groupby('test preparation course')['reading score'].mean().plot(kind='bar',figsize=(12,4),title='Avg reading score by TPC')
plt.xlabel('Test Preparation Course')
plt.ylabel('Avg Score')
plt.subplot(1,3,3)
plot3=data.groupby('test preparation course')['writing score'].mean().plot(kind='bar',figsize=(12,4),title='Avg writing score by TPC')
plt.xlabel('Test Preparation Course')
plt.ylabel('Avg Score')
```

From the plot it is clear that test preparation course helped the students with better performance in the tests.
Those who completed the course performed better than those who didn't

# Score variation by gender

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.boxplot(x=data['gender'],y=data['math score'])

plt.subplot(1,3,2)
sns.boxplot(x=data['gender'],y=data['reading score'])

plt.subplot(1,3,3)
sns.boxplot(x=data['gender'],y=data['writing score'])
```

From the plot it is seen that males showed less variation in math score whereas females showed more variations.
For reading and writing females showed more consistency in their scores compared to male.

# Score variation by test preparation course

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.boxplot(x=data['test preparation course'],y=data['math score'])

plt.subplot(1,3,2)
sns.boxplot(x=data['test preparation course'],y=data['reading score'])

plt.subplot(1,3,3)
sns.boxplot(x=data['test preparation course'],y=data['writing score'])
```

In all the tests those who took the course showed more consistency in their scores than those who didn't.
