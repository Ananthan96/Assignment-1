# Creating a contingency  table

```python
data1=pd.crosstab(data.Region, data.Manager,margins=True)
data1
```

# Defining the table and Interpreting p value.

```python
data2 = [[4, 3, 3], [ 4, 1, 2], [1, 3, 1]]
stat, p, dof, expected = chi2_contingency(data2)
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

From the Chi-Squared test for Independence, since p value greater than 0.05 null hypothesis holds and there is no association between Region and Manager. They are Independent
